COLLECTIONS-CONCEPT


```python
from collections import Counter
```

1.COUNTER

```python
a="aaaabbbbccc"
my_counter=Counter(a)
print(my_counter)
```

```python
a="aaaabbbbccc"
my_counter=Counter(a)
print(my_counter.items())
```

```python
a="aaaabbbbccc"
my_counter=Counter(a)
print(my_counter.keys())
```

```python
a="aaaabbbbccc"
my_counter=Counter(a)
print(my_counter.values())
print(my_counter.most_common(1)[0][0])
```

```python
a="aaaabbbbccc"
my_counter=Counter(a)
print(my_counter)
print(list(my_counter.elements()))
```

NAMEDTUPLE-is an easy to create and light weight object type similar to struct

```python
from collections import namedtuple
```

```python
point=namedtuple('point','x,y')
pt=point(1,-4)#it creates class like point and assign the parameters like x,y
print(pt.x,pt.y)
```

ORDERED DICT-it is a regualar dictionary but they remember the order that items were inserted

```python
from collections import OrderedDict
```

```python
ordered_dict=OrderedDict()
ordered_dict={}
ordered_dict['a']=5
ordered_dict['b']=3
ordered_dict['c']=2
print(ordered_dict)
```

DEFAULTDICT-it is similar to the dictionary container with the only difference that will have default value if the key has not been yet so

```python
from collections import defaultdict
```

```python
d=defaultdict(int)
d['a']=1
d['b']=2
print(d['a'])
print(d['b'])
print(d['c'])
```

```python
d=defaultdict(int)
d=defaultdict(list)
d['a']=1
d['b']=2
print(d['a'])
print(d['b'])
print(d['c'])
```

dqeue-remove elements from both end and very efficently


```python
from collections import deque
```

```python
d=deque()
d.append(1)
d.append(2)
d.appendleft(3)
d.extendleft([5,6,7])
print(d)
d.rotate(1)
print(d)
print(deque)
```

ITERTOOLS-it is a collection of tools for handling iterators


PRODUCT-it combines and give iterations

```python
from itertools import product
a=[1,2]
b=[3,4]
prod=product(a,b,repeat=2)
print(list(prod))
```

permuatations-give possible orderings

```python
from itertools import permutations
a=[1,2,3]
perm=permutations(a,2)
print(list(perm))
```

combinations-given order of combinations

```python
from itertools import combinations
a=[1,2,3,3]
comb=combinations(a,2)
print(list(comb))
```

```python
from itertools import combinations,combinations_with_replacement
a=[1,2,3,4]
comb_wr=combinations_with_replacement(a,2)
print(list(comb_wr))
```

accumlate- default the list will be sum

```python
from itertools import accumulate
a=[5,11,3,2,9]
acc=accumulate(a)
print(a)
print(list(acc))
```

```python
import operator
acc=accumulate(a,func=operator.mul)
ac=accumulate(a,func=max)
print(list(a))
print(list(acc))
```

GROUPBY-it is makes an iterator that return keys and groups from an iterable

```python
from itertools import groupby

```

```python
def smaller_than_3(x):
  return x<3
a=[1,2,3,4]
group_obj=groupby(a,key=smaller_than_3)
for key, value in group_obj:
  print(key,list(value))
```

```python
def smaller_than_3(x):
  return x<3
a=[1,2,3,4]
group_obj=groupby(a,key=lambda x:x<3)
for key, value in group_obj:
  print(key,list(value))
```

COUNT,CYCLE,REPEAT

```python
from itertools import count,cycle,repeat
for i in count(10):
  print(i)
  if i==15:
    break
```

```python
from itertools import count,cycle,repeat
a=[1,2,3]
for i in cycle(a):
  print(i)
```

```python
from itertools import count,cycle,repeat
a=[1,2,3]
for i in repeat(a):
  print(i)
  if i==15:
    break
```
