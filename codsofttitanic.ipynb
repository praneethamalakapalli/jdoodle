HANDLING THE MISSING VALUES BY USING DROPNA(),FILLNA(),REPLACE()

```python
# by using drop funtion we can handling the missing  values
df=titanic.drop(columns='Cabin',axis=1)
df
```

```python
df['Age'].mean()
```

```python
#in the age column we have 177 missing values so we can handle by using fillna() function
df['Age'].fillna(df['Age'].mean(),inplace=True)
```

```python
df.isnull().sum()# by observing the column age before handling the age column have 177 missing values byusing fillna with mean we can observe the age columns have no missing value


```

```python
#caluclate mode of the embarked column
df['Embarked'].mode()
```

```python
df['Embarked'].mode()[0]
```

```python
#by replacing fill na function we can replace the values in the embarked column for handling missing values
df['Embarked'].fillna(df['Embarked'].mode()[0],inplace=True)
```

```python
#by observing before the handling the missing value the embarked columns has missing values by using fillna function and by using mode function we can handle the missing  values
df.isnull().sum()
```

TESTING AND TRAINING

```python
#separting the feautres and target
X=df.drop(columns=['PassengerId','Name','Ticket','Survived'],axis=1)
Y=df['Survived']
```

```python
print(X)
```

```python
print(Y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=3)
```

```python
X_train,Y_train
```


MODEL SELECTION

lOGISTIC REGRESSION  it is one of the type of regression and it is supervised learning techinque in which it handles the categorical values,binary values

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

Accuracy score



```python
from sklearn.metrics import accuracy_score
```

```python
X_pred=model.predict(X_train)
```

```python
X_pred
```

```python
train_data=accuracy_score(Y_train,X_pred)
```

```python
train_data#accuracy score of train_data
```

```python
X_test_pred=model.predict(X_test)
```
